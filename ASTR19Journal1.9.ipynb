```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Use np.random.unifrom to generate 1000 between 0 to 1 
random_num = np.random.uniform(0, 1, 1000)

#Historgram
plt.figure(figsize=(6, 9))
#hist is the historgram function / Required 100 bin
plt.hist(random_num, bins=100, color='purple', edgecolor='cyan')

#axes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title(' Random Thousand Number Histogram')

plt.savefig('random_numbers_histogram.pdf')
plt.show()
```
