```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
#Define Sinx
def sin_funtion(x):
    return np.sin(x)

#Define CosX
def cos_funtion(x):
    return np.cos(x)
```

```python
#Y For SinX & CosX
y_sin = sin_funtion(x)
y_cos = cos_funtion(x)

fig, axes = plt.subplots(1, 2, figsize =(6,9))

#Multipanel plot

#SinX
axes[0].plot(x, y_sin, color='blue')
axes[0].set_title('sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].grid(True)

#CosX
axes[1].plot(x, y_sin, color='red')
axes[1].set_title('cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
axes[1].grid(True)

plt.tight_layout()
plt.savefig('sin_cos_plot.pdf')
plt.show()
```
