```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
#Define Exponent X & Y
def exp_function(x):
    return np.exp(x)
y = exp_function(x)
```

```python
# plot
plt.figure()
plt.plot(x, y, label='y = exp(x)', color='blue')
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.title('Plot of exp(x) vs. x')
plt.legend()
plt.grid(True)

# Save figure to PDF
plt.savefig('exp_plot.pdf')
plt.show()
```
