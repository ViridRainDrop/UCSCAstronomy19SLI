```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Get Parameter
p = 0.5

#use np.random.geometric
Random_number = np.random.geometric(p, 1000)

#Histogram
plt.figure(figsize=(6,9))
plt.hist(Random_number, bins=10, color='cyan', edgecolor='black')

#axes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Geometric Histogram')

plt.savefig('Geometric Histagram.pdf')
plt.show()
```
